## Visualisation

We will be recreating the visualisation done by u/urlander on r/dataisbeautiful (https://www.reddit.com/r/dataisbeautiful/comments/br2hw5/oc_google_searches_for_cancel_hbo_subscription/), but with a few small changes to see how it can be better done.

### imports and data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
cancel = pd.read_csv("multiTimeline-4.csv")
```

```python
search = pd.read_csv("multiTimeline-3.csv")
```

```python
weeks = cancel["Week"]
```

```python
cancel_numbers = cancel["cancel hbo subscription: (United States)"]
```

```python
search_numbers = search["subscribe to HBO: (United States)"]
```

Let us do a quick brief plot to see how things look before tidying up:

```python
plt.plot(weeks, cancel_numbers)
```

Yikes: the exact dates look horrible, so let us instead just use numbers.

```python
len(weeks)
```

```python
week_numbers = np.arange(1, 53)
```

```python
week_numbers
```

```python
plt.figure(figsize=(16, 5))
plt.plot(week_numbers, cancel_numbers, label="Cancellation", color="Red")
plt.plot(week_numbers, search_numbers, label="Starting", color="Blue")
plt.ylabel('Popularity of searches')
plt.xlabel('Weeks starting from 2018-06-03')
plt.title("Searches for starting and cancelled HBO subscription")
plt.legend()
plt.show()
```

There we go!

We now have a clearer representation of how Game of Thrones might have effected subscribing to or cancelling HBO subscriptions. 

The story which we wish to convey is that Game of Thrones is popular enough to dictate how people sign up for HBO. While the visualisation by u\urlander also showed us that it is important, it just might be that proportionally, in a week, everyone would only search for cancelling subscriptions after the shows finale airs - this seems obvious. What we would like to do is show how MUCH Game of Thrones dominates this by collecting search data from over a year to see if it is really only Game of Thrones or if this happens for a lot of other shows. 

The other visualisation only shows cancellations, which might not be the whole picture - we would also like to see how often people would want to subscribe to HBO. Would there be a lot of searches right before Game of Thrones starts, for example?

We can see there are lots of little peaks and valleys - this tells us that it might be the end of other popular shows, but there are certainly two very clear peaks for each of the respective searches. One is right before Game of Thrones started to be aired, and one is right after Game of Thrones aired its finale.

The picture is a lot more clear now - over the year there are a lot of ups and downs based on other popular (but less so than Game of Thrones) shows, but in mid-late April 2019 and in mid-late May 2019, there are tow massive peaks. One would be for when Game of Thrones started, and the other for when Game of Thrones finished.

I used a simple line graph to depict this, as all the information (and the story) can be elegantly and minimalistically depicted this way. I choose the color red for cancellations, and the color blue for searches to subscribe. I spread the graph out a bit so that we can see the appropriate scaling over the year. I use a very minimalist design so as to depict a very clear picture.
